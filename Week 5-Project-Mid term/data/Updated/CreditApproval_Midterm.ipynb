```python
# there are three outliers 'chilren_count','annual_income','family_size'
q_hi = app_rec['CNT_CHILDREN'].quantile(0.999)
q_low = app_rec['CNT_CHILDREN'].quantile(0.001)
app_rec = app_rec[(app_rec['CNT_CHILDREN']>q_low) & (app_rec['CNT_CHILDREN']<q_hi)]
q_hi = app_rec['AMT_INCOME_TOTAL'].quantile(0.999)
q_low = app_rec['AMT_INCOME_TOTAL'].quantile(0.001)
app_rec = app_rec[(app_rec['AMT_INCOME_TOTAL']>q_low) & (app_rec['AMT_INCOME_TOTAL']<q_hi)]
q_hi = app_rec['CNT_FAM_MEMBERS'].quantile(0.999)
q_low = app_rec['CNT_FAM_MEMBERS'].quantile(0.001)
app_rec = app_rec[(app_rec['CNT_FAM_MEMBERS']>q_low) & (app_rec['CNT_FAM_MEMBERS']<q_hi)]
```

## Modelling

### Splitting X and Y:

```python
X = cred_approval_df.iloc[:,1:-1] # X value contains all the variables except labels
y = cred_approval_df.iloc[:,-1]   # y contacins the labels('status')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

### Scaling X_train and X_test

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X_train_scaled = pd.DataFrame(mms.fit_transform(X_train), columns=X_train.columns)
X_test_scaled = pd.DataFrame(mms.transform(X_test), columns=X_test.columns)
# we have now fit and transform the data into a scaler for accurate reading and results.
```

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
X_train_balanced, y_train_balanced = oversample.fit_resample(X_train_scaled, y_train)
X_test_balanced, y_test_balanced = oversample.fit_resample(X_test_scaled, y_test)
# we have addressed the issue of oversampling here
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Assuming your DataFrame is called "cred_approval_df" with independent columns 'feature1', 'feature2', etc.,
# and dependent column 'status'

# Create a figure and axes for subplots
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(25, 25))

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Scatter plots with regression lines/curves of each feature against the target variable
for i, column in enumerate(cred_approval_df.columns):
    if column != 'status':
        if i < 16:  # Check if the index is within the bounds of the axes array
            ax = axes[i]
            sns.scatterplot(x=column, y='status', data=cred_approval_df, ax=ax)
            ax.set_xlabel(column)
            ax.set_ylabel('status')

            # Perform linear regression
            X = cred_approval_df[[column]]
            y = cred_approval_df['status']
            reg_model = LinearRegression()
            reg_model.fit(X, y)

            # Plot the regression line
            ax.plot(X, reg_model.predict(X), color='red', label='Linear Regression')

            # Perform polynomial regression
            poly_features = PolynomialFeatures(degree=2)
            X_poly = poly_features.fit_transform(X)
            poly_reg_model = LinearRegression()
            poly_reg_model.fit(X_poly, y)

            # Plot the polynomial regression curve
            x_range = np.linspace(X.min(), X.max(), 100)
            x_range_poly = poly_features.transform(x_range.reshape(-1, 1))
            ax.plot(x_range, poly_reg_model.predict(x_range_poly), color='blue', label='Polynomial Regression (Degree 2)')

            ax.legend()

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
